```python
import json

#Read data to json.
merged_json = json.loads(merged.to_json())

#Convert to String like object.
json_data = json.dumps(merged_json)
```

```python
from bokeh.io import output_notebook, show, output_file, save, export_png, export_svgs
from bokeh.plotting import figure
from bokeh.models import GeoJSONDataSource, LinearColorMapper, ColorBar
from bokeh.palettes import brewer



#Input GeoJSON source that contains features for plotting.
geosource = GeoJSONDataSource(geojson = json_data)

#Define a sequential multi-hue color palette.
palette = brewer['YlOrRd'][8]

#Reverse color order so that dark blue is highest obesity.
palette = palette[::-1]

#Instantiate LinearColorMapper that linearly maps numbers in a range, into a sequence of colors.
color_mapper = LinearColorMapper(palette = palette, low = 0.0, high = 17)

#Define custom tick labels for color bar.
tick_labels = {'0': '0%', '2': '2%', '3':'3%', '5':'5%', '10':'10%','13':'>13%'}

#Create color bar. 
color_bar = ColorBar(color_mapper=color_mapper, label_standoff=8,width = 20, height = 375,
border_line_color=None,location = (0,0), orientation = 'vertical', major_label_overrides = tick_labels)

#Create figure object.
p = figure(title = 'Total_Co2_emissions_in_Percentage', plot_height = 500 , plot_width = 1000, toolbar_location = None)
p.xgrid.grid_line_color = None
p.ygrid.grid_line_color = None

p.xaxis.visible = None
p.yaxis.visible = None

from bokeh.models import LinearAxis
xaxis = LinearAxis(axis_label="Longitude")
yaxis = LinearAxis(axis_label="Latitude")
p.add_layout(xaxis, 'below')
p.add_layout(yaxis, 'left')

#Font size and alignment of title
p.title.text_font_size = "18pt"
p.title.align = "center"

#Font size of labels and ticks
p.axis.axis_label_text_font_size = "18pt"
p.axis.major_label_text_font_size = "18pt"


#Add patch renderer to figure. 
p.patches('xs','ys', source = geosource,fill_color = {'field' :'Total_percent_Co2', 'transform' : color_mapper},
          line_color = 'black', line_width = 0.25, fill_alpha = 1)

#Specify figure layout.
p.add_layout(color_bar, 'right')

#To save html file into system
#output_file('plot.html', mode='inline')
#save(p)

#export_png(p, filename="plot.png")


#Display figure inline in Jupyter Notebook.
output_notebook()

#Display figure.
show(p)
```
